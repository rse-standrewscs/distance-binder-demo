# Getting Started with dssd

## Distance sampling survey design.

Based on https://examples.distancesampling.org/dssd-getting-started/GettingStarted-distill.html by Laura Marshall

```r
library(dssd)
```

### Setting up the Region

```r
shapefile.name <- system.file("extdata", "TrackExample.shp", package = "dssd")
```

```r
region <- make.region(region.name = "study area",
                      shape = shapefile.name)
```

```r
plot(region)
```

### Creating a Coverage Grid

```r
cover <- make.coverage(region,
                       n.grid.points = 1000)
```

### Defining the Design

```r
design <- make.design(region = region, 
                      design = "systematic",
                      line.length = 1300,
                      edge.protocol = "minus",
                      design.angle = 0,
                      truncation = 2,
                      coverage.grid = cover)
```

### Generating a Survey

```r
transects <- generate.transects(design)
```

```r
transects
```

```r
plot(region, transects)
```

```r
plot(region, transects, covered.area = TRUE)
```
